```python
df = pd.read_csv("data.csv")
```

```python
print(df.shape)
```

```python
x = pd.get_dummies(df.drop(['gameId', 'queueId', 'blueTeamWin'], axis=1))
y = df['blueTeamWin']
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Dropout
from keras.optimizers import SGD, Adam
from sklearn.metrics import accuracy_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train=np.asarray(x_train).astype(int)
y_train=np.asarray(y_train).astype(int)

x_test=np.asarray(x_test).astype(int)
y_test=np.asarray(y_test).astype(int)
```

```python
model = Sequential()
model.add(Dense(units=128, activation='relu', input_dim=x_train.shape[1])) # hidden layer 1 taking in x_train.columns number of inputs aka the number of features
model.add(Dropout(0.2))
model.add(Dense(units=128, activation='relu')) # hidden layer 2
model.add(Dropout(0.1))

model.add(Dense(units=1, activation='sigmoid')) # this is the output layer
```

```python
model.compile(loss='binary_crossentropy', optimizer=Adam(0.001), metrics='accuracy')
```

```python
model.fit(x_train, y_train, epochs=100, batch_size=32)
```

```python
y_hat = model.predict(x_test)
y_hat = [0 if val < 0.5 else 1 for val in y_hat] # list comprehension to convert to binary outcome from continuous values
```

```python
y_hat
```

```python
accuracy_score(y_test, y_hat)
```

```python
model.save("tfmodel3")
```
